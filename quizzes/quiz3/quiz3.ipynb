(b) Recode a numeric colum

<font size = "3">

- Import the "data_raw/races.csv" file
- Recode the "year" column into a new column "year_brackets" with the following categories

$\qquad$ ``` ["1950-1974","1975-1999","2000-onwards"] ```

- Display the new "year_brackets" variable

$\qquad$ HINT: Use the "pd.cut()" command. See Lecture 12 (check the note in the example!)

````python
races_raw = pd.read_csv('data_raw/races.csv')
````

````python
# Write your own code
races_raw = pd.read_csv('data_raw/races.csv')
bins_x = [1950,1975,2000,float('inf')]
labels_x = ["1950-1974",
            "1975-1999",
            "2000-onwards"]
races_raw['year_brackets'] = pd.cut(races_raw['year'],
                                    bins = bins_x,
                                    right = True,
                                    labels = labels_x)
races_raw
# Note: if we set bins_x = [float("-inf"),2500, float("inf")]
#       then intervals are "Less than or equal to 2500" and "Above 2500"
#       float("inf") and float("-inf") represent infinity and negative infinity
#       The "right" command indicates that the right interval is
#       "less than or equal to" or just "less than"
````

(c) Aggregate and query

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Subset the observations for 'constructorId == 3' using ```.query()```
- Obtain the sum of "laps" by "driverId" using ``` .groupby().agg() ```
- Subset the drivers with a sum of laps $\ge 40$
- Store the results of all these operations in a dataframe ```sprint_results_agg ```
- Display the ```sprint_results_agg``` dataframe
- Provide a one-sentence explanation of the information in the ```sprint_results_agg ``` table

````python
# Write your own code
sprint_results_raw = pd.read_csv('data_raw/sprint_results.csv')
sprint_results_raw.query('constructorId == 3')
sprint_results_raw
sprint_results_raw = sprint_results_raw.groupby('driverId').agg(sum_laps = ('laps','sum'))
sprint_results_raw.query('sum_laps >= 40')
sprint_results_agg = pd.DataFrame(sprint_results_raw)
sprint_results_agg
#The information in this table provides a grouped display of drivers who have accumulated over 40 laps in their respective races. 
````

(d) Aggregate and sort

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Obtain the sum of "laps" by "driverId" using ``` .groupby().agg() ```
- Sort the teams in descending order using of the sum of laps using ``` .sort_values() ```
- Display your results

````python
# Write your own code
sprint_results_new = pd.read_csv('data_raw/sprint_results.csv')
sprint_results_agg = sprint_results_new.groupby("driverId").agg(s_laps = ('laps','sum'))
sorted = sprint_results_agg.sort_values(by = "s_laps", ascending = False). reset_index()
sorted
````
